```python

image = imageio.imread('tusur.bmp')
image
```

```python
A = image[:,:,1]
```

```python
%time u,s,v = np.linalg.svd(A, full_matrices = 0)
```

```python
plt.subplots(1,2, figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(A)
plt.title('Зеленый канал ориг. ТУСУРа')
plt.axis('off')
plt.subplot(1,2,2)
rank = 60
plt.imshow(u[:,:rank]@np.diag(s[:rank])@v[:rank,:])
plt.title('Восстановленное')
plt.axis('off')
plt.show()
```

```python

plt.subplots(1,2, figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(u[:,:rank]@np.diag(s[:rank])@v[:rank,:])
plt.title('SVD из 60 компонент')
plt.axis('off')
def rsvd(A, Omega):
    Y = A @ Omega
    Q, _ = np.linalg.qr(Y)
    B = Q.T @ A
    u_tilde, s, v = np.linalg.svd(B, full_matrices = 0)
    u = Q @ u_tilde
    return u, s, v
rank = 60
Omega = np.random.randn(A.shape[1], rank)
%time u,s,v= rsvd(A,Omega)

plt.subplot(1,2,2)
plt.imshow(u[:,:rank]@np.diag(s[:rank])@v[:rank,:])
plt.title('Рандомизированное из 60 компонент')
plt.axis('off')
plt.show()
```

Обычное разложение 76 ms, рандомизированное - 13 ms. Необходимое количество компонент - 60 (выбрано исключительно визуально)
